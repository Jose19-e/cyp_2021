```
#Ejercicio 4.11
print("Ingrese el total de países de Sur, Centro y Norteamérica")
TPS = int ( input("Total de países de america del sur: "))
TPC = int ( input("Total de países de america central: "))
TPN = int ( input("Total de países de america del norte: "))
SUR = [0.0 for x in range(TPS)]
CENTRO = [0.0 for x in range(TPC)]
NORTE = [0.0 for x in range(TPN)]
AMERICA = [0.0 for x in range(TPS + TPC + TPN + 3)]
if TPS >= 1 and TPS <= 30 and TPC >= 1 and TPC <= 30 and TPN >= 1 and TPN <= 30:
  print("Ingrese el nombre de los países de Suramérica: ")
  for i in range(TPS):
    SUR[i] = str(input(f"{i + 1}. "))
      
  print("Ingrese los nombres de los países de Centroamérica: ")
  for i in range(TPC):
    CENTRO[i] = str(input(f"{i + 1}. "))
    
  print("Ingrese el nombre de los países de Norteamérica: ")  
  for i in range(TPN):
    NORTE[i] = str(input(f"{i + 1}. "))
    
  PS = 0
  PC = 0
  PN = 0
  PA = 0 
  while (PS != TPS) and (PC != TPC) and (PN != TPN):
    PA = PA + 1
    if (SUR[PS] < CENTRO[PC]) and (SUR[PS] < NORTE[PN]):
      AMERICA[PA] = SUR[PS]
      PS = PS + 1
    else:
      if CENTRO[PC] < NORTE[PN]:
        AMERICA[PA] = CENTRO[PC] 
        PC = PC + 1
      else: 
        AMERICA[PA] = NORTE[PN] 
        PN = PN + 1
  while (PS != TPS) and (PC != TPC):
    PA = PA + 1
    if SUR[PS] < CENTRO[PC]:
      AMERICA[PA] = SUR[PS]
      PS = PS + 1
    else:
      AMERICA[PA] = CENTRO[PC]  
      PC = PC + 1
  while (PS != TPS) and (PN != TPN):
    PA = PA + 1
    if SUR[PS] < NORTE[PN]:
      AMERICA[PA] = SUR[PS]
      PS = PS + 1
    else:
      AMERICA[PA] = NORTE[PN]  
      PN = PN + 1
  while (PC != TPC ) and (PN != TPN):
    PA = PA + 1
    if CENTRO[PC] < NORTE[PN]:
      AMERICA[PA] = CENTRO[PC]
      PC = PC + 1
    else:
      AMERICA[PA] = NORTE[PN]    
      PN = PN + 1 
  while PS != TPS:
    PA = PA + 1
    AMERICA[PA] = SUR[PS]
    PS = PS + 1
  while PC != TPC:
    PA = PA + 1
    AMERICA[PA] = CENTRO[PC]  
    PC = PC + 1
  while PN != TPN:
    PA = PA + 1
    AMERICA[PA] = NORTE[PN]  
    PN = PN + 1
    
  print("AMÉRICA")
  for i in range (PA):
    print(str(i + 1) + ". " + str(AMERICA[i + 1]))
    
else:
  print("Verifique el número de países ingresados")    


  
 


  
```
